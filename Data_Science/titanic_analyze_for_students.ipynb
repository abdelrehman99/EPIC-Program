#Task 6. What is the most popular family name in the dataset (1 pt)

```python
arr = df["Name"].tolist()

arr2 = []

for s in arr:
    arr2.append(s.split(', ')[1])

s = pd.Series(arr2)

print(s.value_counts().index[0])
```
